```python
from IPython.display import display, HTML
from jinja2 import Template
import cProfile as _cprofile
import pstats as _pstats
from snakeviz import stats as sstats
import numpy as np
```

### Setting up the data
The plot ....

```python
def dot(a, b):
    sum=0
    for i in range(len(a)):
        sum += a[i]*b[i]
    return sum

profiler = _cprofile.Profile()

a = np.arange(16, dtype=np.float32)
b = np.arange(16, dtype=np.float32)
profiler.run("dot(a, b)")
stats = _pstats.Stats(profiler)
table_rows=sstats.table_rows(stats)
callees=sstats.json_stats(stats)
```

```python
# Open our custom template
with open('snakeviz_template.jinja', 'r') as f:
    template = Template(f.read())

html = template.render(table_rows=table_rows, callees=callees)
# Rendering the html as a static file works fine...
open("static.html", 'w').write(html)
# ...but rendering embedded into a notebook yields javascript errors.
HTML(html)
```
